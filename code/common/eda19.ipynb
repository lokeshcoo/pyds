```python
def plot_candle_stick(stock):

    fig = go.Figure(data=[go.Candlestick(x=stock['Date'],
                    open=stock['Open'],
                    high=stock['High'],
                    low=stock['Low'],
                    close=stock['Close'])])

    return fig
```

```python
plot_candle_stick(df)
```
